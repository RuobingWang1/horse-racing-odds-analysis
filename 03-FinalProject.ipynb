## Abstract

The objective of this project is to examine the relationship between the odds given for a horse before the race and the average return rates for horse racing events at the Saga Racecourse in Japan. Guided by the findings of previous studies on the "favorite-longshot bias," the research question here is: do the odds given for a horse before the race relate to a different average return pattern for those with lower odds compared with those with higher odds? The hypothesis for this study is: there is a statistically significant inverse and nonlinear relationship between the odds given for a horse and the average return rates.
To verify this problem, we used a structured data set with more than 120,000 records related to horse racing, including pre-race odds and race results. After completing data cleaning, we computed real returns according to winning results, greatly diminishing high noise levels generated by individual race results. Then, we researched the relationship between odds and average returns using various methods, such as descriptive statistics, data visualization, logarithmic transformation, and Pearson/Spearman correlation tests.
It is observed that there is a significant negative correlation between the betting odds and the average return rate, as the odds increase, the average return rate decreases in an overall manner, as observed from the result obtained after performing the binning analysis on the data with the help of the logarithmic scale. From the result obtained, the alternative hypothesis proposed in the paper is confirmed, as the data obtained from the betting market of SAGA Racecourse shows the presence of the "favorite-longshot bias" theory, as the horses with low odds are observed to have the best performance in terms of the average return rate. Thus, the paper shows the presence of a systematic relationship between the odds and the return rates in the betting market of SAGA Racecourse, which could indicate the possible presence of the favorite-longshot bias in the betting market of SAGA Racecourse.

## Background and Prior Work

Horse racing is one of the oldest organized sports in the world, with a documented history dating back to ancient civilizations such as Greece, Rome, and China. Morden horse racing developed in Britain during the 17th and 18th centuries and later spread to the United States, where it was closely tied to gambling and probability-based decision making.1 Today, horse racing generates large-scale structured data, including race conditions, horse characteristics, jockey performance, and betting odds, making it a natural domain for quantitative analysis and predictive modeling.

Beyond its historical and cultural significance, horse racing has served as an important real-world testbed for studying human decision-making under uncertainty for a long time. Betting markets associated with horse races have frequently been analyzed to understand whether markets efficiently incorporate available information.

Prior economic research has shown that while betting odds often approximate true winning probabilities, systematic biases, such as the “favorite–longshot bias,” where longshots are overbet and favorites are underbet. This persists across races and contexts.2 These findings suggest that even in environments with rich information, human judgment and behavior can deviate from purely rational models.

While prior research has primarily examined large and nationally prominent racing markets, our project focuses specifically on data from the Saga Racecourse in Japan. This makes it a suitable case study for investigating whether well documented behavioral patterns in betting markets，such as the favorite longshot bias also appear in smaller or regional contexts. Because Saga operates under the same pari-mutuel betting system as larger national tracks, it provides a comparable institutional setting for evaluating whether similar decision making patterns emerge. By studying Saga, we are able to test whether these biases are robust across different market scales.

Our research question directly engages with the favorite longshot bias by examining whether betting odds systematically misestimate true winning probabilities in this dataset. Specifically, we analyze the relationship between implied probabilities derived from betting odds and actual race outcomes to determine whether longshots are overvalued and favorites undervalued, as suggested by prior economic research. Through the analysis, our project not only applies predictive modeling techniques to real world data but also evaluates whether deviations from rational betting behavior persist in this particular racing environment.

```python
# --- Data Wrangling ---
# 1. load data
import pandas as pd
import numpy as np
data = pd.read_csv('data/00-raw/saga_entries_odds.csv', low_memory = False)

# 2. inspection
print('\nGet a general view of values in the dataset: \n')
display(data.info())
print('\nCheck the number of nulls in each column: \n')
display(data.isna().sum())

# 3. cleaning
# A null rank often happens because a horse or jockey conducts a rule violation,
# or the horse withdrew from the race because of an accident.
# Thus, we will rank them as last, and fill time as the last horse + 0.1s.
# Function to fill null ranks as last place in each race
def fill_null_ranks(group):
    # Get the maximum rank in this race (excluding nulls)
    max_rank = group['rank'].max()
    # Get the maximum time in this race (excluding nulls)
    max_time = group['time'].max()

    # Fill null ranks with max_rank + 1
    group['rank'] = group['rank'].fillna(max_rank + 1)
    # Fill null time with max_time + 0.1
    group['time'] = group['time'].fillna(max_time + 0.1)

    return group
# Group by race_id and apply the function
data = data.groupby('race_id', group_keys=False).apply(fill_null_ranks)
# Verify the changes
print('Check the number of nulls in each column again:')
display(data.isna().sum())

# 4. summary statistics
import seaborn as sns
import matplotlib.pyplot as plt

# 4.1. Distribution of Odds
plt.figure(figsize=(8, 5))
sns.histplot(
    data['win_odds'],
    bins=50,
    kde=True,
    color='darkorange'
)
plt.xlabel("Winning Odds")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.3)
plt.show()

# 4.2. Market Probabilities
# We will investigate the market's prediction compared to real circumstances.
# Thus we have to convert odds to probabilities.
# Note that because 20% of bets are taken by the racecourse for taxation, maintenance and other reasons,
# The market probabilities calculated by inverse of odds does not sum to 1, but sums to 1.25 per race.
data['market_probs'] = 1 / data['win_odds']
# we also plot its distribution:
plt.figure(figsize=(8, 5))
sns.histplot(
    data['market_probs'],
    bins=50,
    kde=True,
    color='blue'
)
plt.xlabel("Market Probabilities")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.3)
plt.show()

# 4.3. Log Probabilities
# Now the data is very skewed.
# If we imagine the model of a horse racing to be that the expected probability of winning of the best horse is p,
# then the probability of the second popular horse to win can be estimated by p(1-p), the third p(1-p)(1-p),
# so if we take the logarithm of probabilities, it is expected that we can convert it to a better shape.
data['market_log'] = np.log(data['market_probs'])
plt.figure(figsize=(8, 5))
sns.histplot(
    data['market_log'],
    bins=50,
    kde=True,
    color='red'
)
plt.xlabel("Log Probabilities")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.3)
plt.show()

# 4.4. Residuals Versus Odds
# Now we want to see the difference between observed probabilities and market_predicted probabilities.
data['residuals'] = data['market_probs'] - (data['rank'] == 1.0).astype(int)
# a positive residual means underestimation, while a negative residual means overestimation.
# We plot the relationship between residuals and market probs to see if favorites are underestimated,
# and longshots are overestimated.
# Note that because 20% of bets are taken by the racecourse for taxation, maintenance and other reasons,
# The market probabilities calculated by inverse of odds does not sum to 1, but sums to 1.25 per race.
# So in general, all horses are overestimated.
# First we bin market probabilities into 30 groups (the number is arbitrary)
plt.figure(figsize=(9, 6))
sns.lineplot(
    data=data,
    x='market_probs',
    y='residuals',
    estimator='mean',
    color='purple',
    lw=2,
    label='Mean residual per bin'
)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title("Residuals vs Market Probabilities (Binned)")
plt.xlabel("Market Probabilities (1 / Win Odds)")
plt.ylabel("Mean Residual (Market Prob - Actual Outcome)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# --- Data Processing ---
data['unique_id'] = (data['race_id'].astype(str) + '_' + data['horse_no'].astype(str)).astype(int)
data['win'] = (data['rank'] == 1.0).astype(int)
data['win_return'] = data['win'] * data['win_odds']
data[['unique_id','win_odds','win','win_return']].dropna().to_csv('data/02-processed/saga_odds.csv')
```

```python
# 0. import models and data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.metrics import brier_score_loss
data_new = pd.read_csv('data/02-processed/saga_odds.csv')
```

#### 2. Binning Odds for Aggregated Analysis

Individual race outcomes are extremely noisy because each bet either results 
in a win or a loss. To reduce noise and reveal underlying patterns, we group 
the betting odds into bins.

By averaging outcomes within bins, we can observe the relationship between 
betting odds and average returns more clearly. 

```python
# 2. use bins for better visualization and simpler analysis
# also removes noise, since actual return is always 0 or >1
# while 'expected' return is always between 0 and 1
# so we have to bin values to make a rational analysis

# save a copy of original data just in case
data_new_copy = data_new.copy()
data_new_copy['win_odds_bin'] = pd.qcut(
    data_new_copy['win_odds'],
    q=30,         # Arbitrary number of EDA data points
    # there are 129439 data points in total, which is enough
    # qcut ensures that bin sizes are similar
    labels=False, # Use integer labels
    duplicates='drop'
)

# Calculate the average win_odds and average win_return for each bin
binned_odds_return = data_new_copy.groupby('win_odds_bin', observed=False).agg(
    average_win_odds=('win_odds', 'mean'),
    average_win_return=('win_return', 'mean')
).reset_index()

display(binned_odds_return.head())
```

#### 3. Statistical Relationship Between Odds and Returns

We next examine the statistical relationship between betting odds and 
realized returns. Several statistical measures are used to evaluate 
this relationship.

First, we compute the Pearson correlation coefficient to measure the 
linear relationship between average odds and average returns.

We also calculate Spearman's rank correlation to evaluate the 
monotonic relationship between the variables without assuming linearity.

```python
# 1. calculate R and p using scipy.stats
correlation_coefficient, p_value = pearsonr(
    binned_odds_return['average_win_odds'],
    binned_odds_return['average_win_return']
)

print(f"Pearson Correlation Coefficient: {correlation_coefficient:.4f}")
print(f"P-value: {p_value:.8f}")
```

```python
# 2. calculate Spearman's Rank Correlation to evaluate effect size without assuming linearity
from scipy.stats import spearmanr

spearman_corr, spearman_p_value = spearmanr(
    binned_odds_return['average_win_odds'],
    binned_odds_return['average_win_return']
)

print(f"Spearman's Rank Correlation Coefficient: {spearman_corr:.4f}")
print(f"P-value for Spearman's correlation: {spearman_p_value:.8f}")
```

#### 5. Relationship Between Odds and Returns

```python
# 2. relationship visualization (linear with logarithmic x scale)
plt.figure(figsize=(10, 7))

# regression
sns.regplot(
    data=binned_odds_return, x='average_win_odds', y='average_win_return', ci=95,
    scatter_kws={'s': 80, 'alpha': 0.7, 'edgecolor': 'black', 'color': 'lightcoral'},
    line_kws={'color': 'darkblue', 'linestyle': '-', 'lw': 2, 'label': 'Regression Line'}
)

# labels
plt.title('Average Win Return vs. Average Win Odds (Linear Scale)')
plt.xlabel('Average Win Odds')
plt.ylabel('Average Win Return')
plt.xscale('log')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

#### 6. Log-Scale Analysis

Because the distribution of betting odds is highly skewed, we also 
examine the relationship between odds and returns on a logarithmic scale.

Logarithmic transformation reduces skewness and helps reveal patterns 
that may not be visible under the original scale.

### Primary Analysis
Here we generate a report of outcomes, using bins to convert discrete values of return (0 or odds) to a local average that could be investigated as local expectation. This could potentially be used to verify our hypothesis:

```python
import warnings
warnings.filterwarnings('ignore')
# suppress warnings for better display

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, ttest_ind
from statsmodels.nonparametric.smoothers_lowess import lowess
import statsmodels.api as sm
from IPython.display import display, Markdown

%matplotlib inline

# Load Data and ensure data integrity
df = pd.read_csv('data/02-processed/saga_odds.csv').dropna(subset=['win_odds', 'win', 'win_return'])

# bin odds
df['odds_bin'] = pd.qcut(df['win_odds'], q=30, labels=False, duplicates='drop')

# summarize binned data and compute confidence interval
grp = df.groupby('odds_bin').agg(
    avg_odds=('win_odds', 'mean'),
    avg_return=('win_return', 'mean'),
    win_rate=('win', 'mean'),
    count=('win', 'size'),
    return_sem=('win_return', lambda x: x.std(ddof=1) / np.sqrt(x.size) if x.size > 1 else 0.0)
).reset_index()
grp['return_ci95'] = 1.96 * grp['return_sem']

# compute pearson and spearman r to test correlation
pearson_r = spearman_r = pearson_p = spearman_p = np.nan
if grp['avg_odds'].nunique() > 1 and grp['avg_return'].nunique() > 1:
    try:
        pearson_r, pearson_p = pearsonr(grp['avg_odds'], grp['avg_return'])
    except Exception:
        pearson_r = pearson_p = np.nan
    try:
        spearman_r, spearman_p = spearmanr(grp['avg_odds'], grp['avg_return'])
    except Exception:
        spearman_r = spearman_p = np.nan

# --- Regression: model average return as a function of log(odds) ---
# 1) OLS on binned averages (more stable, reduces per-horse noise)
reg_binned = None
binned_slope = binned_intercept = binned_p = binned_r2 = np.nan
valid = grp['avg_odds'] > 0
if valid.sum() >= 2:
    Xb = np.log(grp.loc[valid, 'avg_odds'])
    yb = grp.loc[valid, 'avg_return']
    Xb_const = sm.add_constant(Xb)
    try:
        reg_binned = sm.OLS(yb, Xb_const).fit()
        binned_intercept = float(reg_binned.params[0])
        binned_slope = float(reg_binned.params[1])
        binned_p = float(reg_binned.pvalues[1])
        binned_r2 = float(reg_binned.rsquared)
    except Exception:
        binned_intercept = binned_slope = binned_p = binned_r2 = np.nan

# 2) OLS on per-horse returns (noisy but uses all data)
per_slope = per_intercept = per_p = per_r2 = np.nan
mask_pos = df['win_odds'] > 0
if mask_pos.sum() >= 10:
    Xp = np.log(df.loc[mask_pos, 'win_odds'])
    yp = df.loc[mask_pos, 'win_return']
    Xp_const = sm.add_constant(Xp)
    try:
        reg_per = sm.OLS(yp, Xp_const).fit()
        per_intercept = float(reg_per.params[0])
        per_slope = float(reg_per.params[1])
        per_p = float(reg_per.pvalues[1])
        per_r2 = float(reg_per.rsquared)
    except Exception:
        per_intercept = per_slope = per_p = per_r2 = np.nan

# T-test: compare returns for favorites vs longshots (using bin groups)
n_bins = grp['odds_bin'].nunique()
if n_bins >= 10:
    fav_cut = int(max(1, n_bins * 0.10))
    fav_bins = grp.nsmallest(fav_cut, 'avg_odds')['odds_bin'].tolist()
    long_bins = grp.nlargest(fav_cut, 'avg_odds')['odds_bin'].tolist()
else:
    fav_bins = grp.nsmallest(1, 'avg_odds')['odds_bin'].tolist()
    long_bins = grp.nlargest(1, 'avg_odds')['odds_bin'].tolist()

fav_df = df[df['odds_bin'].isin(fav_bins)]
long_df = df[df['odds_bin'].isin(long_bins)]
try:
    t_stat_return, t_p_return = ttest_ind(fav_df['win_return'].dropna(), long_df['win_return'].dropna(), equal_var=False)
except Exception:
    t_stat_return = t_p_return = np.nan

# display summary metrics using markdown
display(Markdown(f"""
**Summary metrics**

- Binned Pearson r (avg_odds vs avg_return): {pearson_r:.4f} (p={pearson_p:.3g})  
A Pearson r close to -1 strong negative linear correlation across bin averages.
- Binned Spearman r: {spearman_r:.4f} (p={spearman_p:.3g})  
A negative Spearman r strong negative monotonic (rank) relationship across bins.
- Binned OLS: avg_return ~ a + b * log(avg_odds): slope={binned_slope:.4f}, intercept={binned_intercept:.4f}, p={binned_p:.3g}, R²={binned_r2:.4f}  
A one-unit increase in log(avg_odds) is associated with a ~{binned_slope:.4f} change in average return. R² ≈ {binned_r2:.4f} implies that log-odds explains most of variance across bins.
- Per-horse OLS: win_return ~ a + b * log(win_odds): slope={per_slope:.4f}, intercept={per_intercept:.4f}, p={per_p:.3g}, R²={per_r2:.4f}  
The slope is similar to the binned slope, but low R value implies that log(odds) is not accurate on an individual, case-by-case basis.
- t-test (favorites vs longshots: win_return): t={t_stat_return:.4f}, p={t_p_return:.3g}  
Low p value indicates that the mean win_return in the favorite bins is significantly different from that of the longshot bins;
Positive t value implies that favorites have higher average return than longshots.
"""))

sns.set(style='whitegrid')

# --- 1) Binned avg win return vs avg odds (plot with LOWESS and binned OLS fit) ---
display(Markdown('### 1) Binned average win return vs average odds'))
plt.figure(figsize=(8, 6))
plt.errorbar(grp['avg_odds'], grp['avg_return'], yerr=grp['return_ci95'],
             fmt='o', alpha=0.8, label='Binned means (95% CI)')
# LOWESS smooth
valid = grp['avg_odds'] > 0
smoothed = lowess(grp.loc[valid, 'avg_return'], grp.loc[valid, 'avg_odds'], frac=0.4)
plt.plot(smoothed[:, 0], smoothed[:, 1], color='red', label='LOWESS smooth')
# plot binned OLS prediction if available
if not np.isnan(binned_slope):
    x_smooth = np.logspace(np.log10(grp.loc[valid, 'avg_odds'].min()), np.log10(grp.loc[valid, 'avg_odds'].max()), 200)
    y_pred = binned_intercept + binned_slope * np.log(x_smooth)
    plt.plot(x_smooth, y_pred, color='orange', linestyle='--', label='Binned OLS fit')

plt.xscale('log')
plt.xlabel('Average Win Odds (log scale)')
plt.ylabel('Average Win Return')
plt.title('Fig. 1 — Binned average win return vs average odds (with 95% CI)')
plt.legend()
plt.tight_layout()
plt.show()
# explanation
display(Markdown('*Fig. 1 shows how average win return varies with odds; the dashed line is an OLS fit on binned averages.*'))

# --- 2) Win rate by odds bin (secondary) ---
display(Markdown('### 2) Win rate by odds bin (secondary) for reference'))
plt.figure(figsize=(6, 4))
sns.lineplot(data=grp.sort_values('avg_odds'), x='avg_odds', y='win_rate', marker='o')
plt.xscale('log')
plt.xlabel('Average Win Odds (log scale)')
plt.ylabel('Win rate')
plt.title('Fig. 2 — Win rate by odds bin (secondary)')
plt.tight_layout()
plt.show()
# explanation
display(Markdown('*Fig. 2 shows raw win probability by odds (secondary to the return analysis).*'))

# t-test style comparison between favs and longs using return metric (already computed earlier)

# Favorites / Longshots numeric summary
fav_tbl = pd.Series({
        'avg_odds': fav_df['win_odds'].mean(),
        'avg_return': fav_df['win_return'].mean(),
        'win_rate': fav_df['win'].mean(),
        'count': len(fav_df)
    })
long_tbl = pd.Series({
        'avg_odds': long_df['win_odds'].mean(),
        'avg_return': long_df['win_return'].mean(),
        'win_rate': long_df['win'].mean(),
        'count': len(long_df)
    })
# Format values
def format_summary(series):
    s = series.copy()
    s['win_rate'] = f"{s['win_rate']*100:.1f}%"
    s['avg_odds'] = f"{s['avg_odds']:.4f}"
    s['avg_return'] = f"{s['avg_return']:.4f}"
    s['count'] = int(s['count'])
    return s
fav_tbl_fmt = format_summary(fav_tbl)
long_tbl_fmt = format_summary(long_tbl)
# display results
display(Markdown('**Favorites summary statistics:**'))
display(fav_tbl_fmt)
display(Markdown('**Longshots summary statistics:**'))
display(long_tbl_fmt)
```

### Analysis 2: Adjusted R
The weighted correlation was calculated using the number of races in each bin as weights, so bins with more observations contribute proportionally more to the correlation. A value that is close to -1 implies strong negative correlation.

```python
display(Markdown('### 3) Weighted correlation summary (strength of negative relationship)'))
# weights = number of races in each bin
weights = grp['count']
x = grp['avg_odds']
y = grp['avg_return']
# Weighted covariance / correlation formula
def weighted_corr(x, y, w):
    w_sum = np.sum(w)
    x_w = np.sum(w * x) / w_sum
    y_w = np.sum(w * y) / w_sum
    cov_w = np.sum(w * (x - x_w) * (y - y_w)) / w_sum
    var_x_w = np.sum(w * (x - x_w)**2) / w_sum
    var_y_w = np.sum(w * (y - y_w)**2) / w_sum
    return cov_w / np.sqrt(var_x_w * var_y_w)
weighted_r = weighted_corr(x, y, weights)
# display results
display(Markdown(f"*Weighted Pearson correlation of average return vs odds (weighted by bin count):* **{weighted_r:.4f}**"))
# if the weighted r is also negative, this supports our hypothesis
if weighted_r < 0:
    display(Markdown("Bins with more entries contribute proportionally more to the adjusted correlation than small bins, further confirming the strong negative relationship."))
```
