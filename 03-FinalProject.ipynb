## Discusison and Conclusion

In this analysis, we investigated the relationship between betting odds and realized returns using the Saga Racecourse dataset. By exploring and analyzing the data, it showed that the distribution of win odds is highly right-skewed. Most horses having relatively low odds and a long tail of high-odds longshots. To better analyze this relationship, betting odds were grouped into bins and average returns were calculated within each bin. The results revealed a strong negative relationship between betting odds and average returns. Pearson’s correlation coefficient between average odds and average returns was −0.8258 (p < 0.001), while Spearman’s rank correlation coefficient was −0.7112 (p < 0.001), indicating a statistically significant negative association. By viewing both linear and logarithmic scales, we confirmed that as betting odds increase, average returns decrease. These findings suggest that favorites tend to produce relatively higher returns compared to longshots, providing evidence consistent with the favorite–longshot bias observed in betting markets.

The results of our analysis can be interpreted in light of the prior research discussed in the Background section. Previous economic studies of betting markets have documented the favorite–longshot bias, where bettors tend to overbet longshots and undervalue favorites, leading to lower expected returns on longshot bets. Our analysis extends this line of work by testing whether the same pattern appears in a regional betting market, specifically the Saga Racecourse in Japan. Although Saga is smaller than the large national markets typically studied in prior literature, it operates under the same pari-mutuel betting system. Our findings show that higher odds are associated with lower average returns and that longshots yield significantly worse returns than favorites. This suggests that the behavioral patterns identified in earlier studies are not limited to large racing markets but can also appear in smaller regional contexts such as Saga.

Taken together, the different analyses consistently point to the same overall pattern. Both the Pearson and Spearman correlations show a strong negative relationship between odds and average returns, indicating that as horses become less favored (i.e., their odds increase), the returns to bettors tend to decrease. The OLS regression on binned averages further confirms this relationship, with a negative slope showing that increases in log(odds) are associated with lower average returns. In addition, the t-test comparing favorites and longshots shows a statistically significant difference in mean returns between these two groups, with favorites producing higher returns on average. When considered together, these results provide multiple forms of statistical evidence supporting the same conclusion: returns decline as odds increase, consistent with the presence of the favorite–longshot bias in this dataset.

Some results may initially appear confusing. For example, the regression performed at the individual horse level produces a very low R² value, suggesting that odds explain little of the variation in single race outcomes. However, this does not contradict the broader findings. Individual race results are highly noisy and influenced by many unpredictable factors, such as race conditions, horse performance on a given day, and random chance. Because of this noise, odds alone cannot strongly predict the return from a single bet. When the data are aggregated into bins, however, random variation is reduced and the underlying relationship between odds and returns becomes clearer. Another potentially confusing observation is that even favorites still have average returns below one, meaning bettors lose money on average. This reflects the built-in house edge of pari-mutuel betting markets. Therefore, our findings do not imply that betting on favorites is profitable, but rather that betting on longshots tends to be even less favorable, which is consistent with the behavioral bias identified in prior research.

Despite these findings, we think there are still several limitations that should be considered. First, the dataset only includes races from Saga Racecourse, which may limit the generalizability of the results to other racecourses or betting markets. Betting behavior and market efficiency can vary across locations, leagues, and betting systems, so the dataset may represent a biased sample of the broader horse racing market. Additionally, although the dataset contains a relatively large number of observations, it is still limited to the races included in the available records. More data from additional racecourses or longer time periods would increase confidence in the results. In addition, this analysis focuses on the overall relationship between betting odds and returns using the available data, rather than examining how this relationship changes over time. If betting markets evolve due to changes in bettor behavior, new technologies, or improved information, the relationship between odds and returns may also change. Therefore, analyzing data across multiple time periods could provide deeper insight into whether the observed bias remains consistent over time.
   
If this investigation were to continue outside this class, we may extend the analysis through several directions. First, include data from multiple racecourses.
